```python
import pandas as pd

# ‚úÖ Use read_csv instead of read_excel
df = pd.read_csv(r"C:\Users\9a33n\OneDrive\Documents\netflix_titles.csv")

# 1Ô∏è‚É£ Identify and handle missing values
df.fillna("Unknown", inplace=True)

# 2Ô∏è‚É£ Remove duplicate rows
df.drop_duplicates(inplace=True)

# 3Ô∏è‚É£ Standardize text values
text_cols = df.select_dtypes(include="object").columns
for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()

# 4Ô∏è‚É£ Convert date formats to consistent type (dd-mm-yyyy)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')

# 5Ô∏è‚É£ Rename column headers to be clean and uniform
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# 6Ô∏è‚É£ Check and fix data types
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')

# üíæ Save cleaned dataset
df.to_csv(r"C:\Users\9a33n\OneDrive\Documents\netflix_titles_cleaned.csv", index=False)

print("‚úÖ Cleaned dataset saved successfully as 'netflix_titles_cleaned.csv'")
```
